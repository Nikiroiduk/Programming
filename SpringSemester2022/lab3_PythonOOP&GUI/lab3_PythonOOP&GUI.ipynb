<br>Дополнить класс таким образом, чтобы выполнялся следующий код:<br>
<code>frac = Fraction(7, 2)</code><br>
<code>print(-frac)</code> # выводит -7/2<br>
<code>print(~frac)</code> # выводит 2/7<br>
<code>print(frac**2)</code> # выводит 49/4<br>
<code>print(float(frac))</code> # выводит 3.5<br>
<code>print(int(frac))</code> # выводит 3


```python
class Fraction(object):
    def __init__(self, num, den):
        self.__num = num
        self.__den = den
        self.reduce()
    def str(self):
        return "%d/%d" % (self.__num, self.__den)
    def reduce(self):
        g = Fraction.gcd(self.__num, self.__den)
        self.__num /= g
        self.__den /= g
    @staticmethod
    def gcd(n, m):
        if m == 0:
            return n
        else:
            return Fraction.gcd(m, n % m)
    
    def __neg__(self):
        return "%d/%d" % (-self.__num, self.__den)

    def __invert__(self):
        return "%d/%d" % (self.__den, self.__num)

    def __pow__(self, other):
        return "%d/%d" % (self.__num ** 2, self.__den ** 2)
    
    def __float__(self):
        return round(self.__num / self.__den, 2)

    def __int__(self):
        return int(self.__num / self.__den)

    def __str__(self):
        return "%d/%d" % (self.__num, self.__den)
        

frac = Fraction(7, 2)
print("print(frac) = ", frac)
print("print(-frac) = ", -frac)
print("print(~frac) = ", ~frac)
print("print(frac**2) = ", frac**2)
print("print(float(frac)) = ", float(frac))
print("print(int(frac)) = ", int(frac))
```

начинаются с большой буквы). Например, tag() для книги с названием 
‘War and Peace’ вернет список тегов [‘War’, ‘Peace’]. Реализуйте классы 
таким образом, чтобы корректно выполнялся следующий код:<br>
<code>lib = Library(1, ’51 Some str., NY’)<br>

&nbsp;&nbsp;&nbsp;&nbsp;print(book.tag())</code>

```python
class Book(object):
    def __init__(self, author, name, year):
        self.__name = name
        self.__author = author
        self.__year = year
    
    def __str__(self):
        tmp = self.__author
        tmp = tmp.split(" ")
        tmp = tmp[0][0] + '. ' + tmp[1]
        return "{} \'{}\'".format(tmp, self.__name)
    
    @property
    def tag(self):
        tmp = self.__name
        tmp = tmp.split(" ")
        return [word for word in tmp if word[0].isupper()]

class Library(Book):
    def __init__(self, adress, number):
        self.__adress = adress
        self.__number = number
    __books = []

    def __iadd__(self, other):
        self.__books.append(other)
        return self

    def __str__(self):
        return "{} {}".format(self.__adress, self.__number)

    def __iter__(self):
        return iter(self.__books)

lb = Library("51 st. of Someone", 911)
lb += Book("Leo Tolstoy", "War and Peace part", 1867)
for book in lb:
    print(book)
    print(book.tag)
```

```python
import sys
import os
import re
from datetime import datetime
from PyQt5.QtCore import Qt
from PyQt5.QtWidgets import QApplication, QLabel, QMainWindow, QAction, QMenu, QFileDialog, QMessageBox, QStatusBar, QWidget, QHBoxLayout, QTextEdit

class Window(QMainWindow):

    def __init__(self, parent=None):
        global log
        super().__init__(parent)
        self.setWindowTitle("Meh")
        self.resize(800, 600)
        self.statusBar = QStatusBar()
        self.setStatusBar(self.statusBar)
        self.actionStatusLabel = QLabel(self)
        self.actionStatusLabel.width = 550
        self.actionStatusLabel.setMaximumWidth(550)
        self.sizeStatusLabel = QLabel(self)
        self.sizeStatusLabel.width = 250
        self.sizeStatusLabel.setMaximumWidth(250)
        widget = QWidget()
        widget.setLayout(QHBoxLayout())
        widget.layout().addWidget(self.actionStatusLabel)
        widget.layout().addWidget(self.sizeStatusLabel)
        self.statusBar.addWidget(widget, 1)
        self.statusBar.addWidget(QWidget())
        self.line = QTextEdit(self)
        self.centralWidget = self.line
        self.setCentralWidget(self.centralWidget)
        self._createMenuBar()
        flag = False
        for filename in os.listdir(os.getcwd()):
            if (filename == "script.log"):
                flag = True
                with open("script.log", "r") as file:
                    lines = file.readlines()
                    for line in lines:
                        log += line
                break
        if not (flag):
            showDialogOk()

    def _createMenuBar(self):
        menuBar = self.menuBar()
        fileMenu = QMenu("&File", self)
        menuBar.addMenu(fileMenu)
        logMenu = menuBar.addMenu("&Log")
        open = QAction("Open...", self)
        export = QAction("Export...", self)
        addLog = QAction("Add log", self)
        check = QAction("Check", self)
        fileMenu.addAction(open)
        logMenu.addAction(export)
        logMenu.addAction(addLog)
        logMenu.addAction(check)
        menuBar.triggered[QAction].connect(self.processtrigger)
    
    def processtrigger(self, q):
        global log
        global lastAction
        if (q.text() == "Open..."):
            try:
                fname = QFileDialog.getOpenFileName(self, "Open File")
                with open(fname[0], 'r') as file:
                    self.sizeStatusLabel.setText('{:,}'.format(os.path.getsize(fname[0])).replace(',', ' ') + " bytes")
                    try:
                        lines = file.readlines()
                        i = 0
                        res = ""
                        for item in lines:
                            i += 1
                            tmp = re.findall("[(]\d{3}[)]\d{3}[-]\d{2}[-]\d{2}|[(]\d{3}[)]\d{3}\d{2}\d{2}", item)
                            for elem in tmp:
                                j = item.index(elem) + 1
                                res += "Line {}, position {}: found '{}'".format(i, j, elem) + "\n"
                        self.line.setText(res)
                        addLogData(str(fname[0]) + " was oppened")
                    except:
                        addLogData("An error has occurred while reading file")
            except:
                addLogData("An error has occurred")
            self.actionStatusLabel.setText(lastAction)
        if(q.text() == "Export..."):
            try:
                name = QFileDialog.getSaveFileName(self, "Save File")
                file = open(str(name[0]),'w')
                file.write(log)
                file.close()
                addLogData(str(name[0]) + " was saved")
            except:
                addLogData("An error has occurred")
            self.actionStatusLabel.setText(lastAction)
        if(q.text() == "Add log"):
            try:
                fname = QFileDialog.getOpenFileName(self, "Open File")
                with open(fname[0], 'r') as file:
                    try:
                        lines = file.readlines()
                        for line in lines:
                            log += line
                        addLogData(str(fname[0]) + " was oppened and succesfully apended to log file")
                    except:
                        addLogData("An error has occurred while adding text in log")
            except:
                addLogData("An error has occurred")
            self.actionStatusLabel.setText(lastAction)
        if(q.text() == "Check"):
            if (showDialogYesNo() == QMessageBox.Yes):
                addLogData("Log data was displayed")
                self.line.setText(log)
                self.actionStatusLabel.setText(lastAction)

lastAction = ""
log = ""
def addLogData(text):
    global log 
    global lastAction
    lastAction = text
    print(text + " " + str(datetime.now()))
    log += "\n" + text + " " + str(datetime.now())
    with open("script.log", "w") as file:
        file.write(log)

def OkBtn(i):
    addLogData(i.text() + " button is pressed")
    with open('script.log', 'w') as file:
        file.write("")

def showDialogYesNo():
    msg = QMessageBox()
    msg.setIcon(QMessageBox.Warning)
    msg.setText("Are you sure?")
    msg.setInformativeText("Are you sure you want to display the log?\nDisplaying it will delete existing information")
    msg.setWindowTitle("Warning")
    msg.setStandardButtons(QMessageBox.Yes | QMessageBox.No)
    return msg.exec_()

def showDialogOk():
    msg = QMessageBox()
    msg.setIcon(QMessageBox.Warning)
    msg.setText("Log file was not found!")
    msg.setInformativeText("script.log file was not found.\nProgram will create it automaticaly")
    msg.setWindowTitle("Warning")
    msg.setStandardButtons(QMessageBox.Ok)
    msg.buttonClicked.connect(OkBtn)
    retval = msg.exec_()

if __name__ == "__main__":
    app = QApplication(sys.argv)
    win = Window()
    win.show()
    app.exec_()
```

#### Задание №4
Напишите простой класс StringFormatter для форматирования строк со 
следующим функционалом:<br>
– удаление всех слов из строки, длина которых меньше n букв;<br>
– замена всех цифр в строке на знак «*»;<br>
– вставка по одному пробелу между всеми символами в строке;<br>
– сортировка слов по размеру;<br>
– сортировка слов в лексикографическом порядке.<br>
Примечание. Разделители слов можно задавать отдельно. По 
умолчанию в качестве разделителя принимается только символ 
пробела.

```python
import re

class StringFormater():

    @staticmethod
    def deleteWords(text, i):
        list = text.split()
        list = [item for item in list if len(item) < i]
        return " ".join(list)
    
    @staticmethod
    def deleteNumbers(text, i):
        for ele in text:
            if ele.isdigit():
                text = text.replace(ele, i)
        return text

    @staticmethod
    def addSpace(text):
        list = text.split()
        return "  ".join(list)
    
    @staticmethod
    def sortWordsbySize(text):
        return " ".join(sorted(text.split(), key=len))

    @staticmethod
    def sortWordsbyAlphabet(text):
        return " ".join(sorted(text.split()))


sf = StringFormater()
print(sf.deleteWords("Test line of text to probably test this class and some more words just in case", 5))
print(sf.deleteNumbers("Test023 494 8234 1re words just in case", "*"))
print(sf.addSpace("Test line of text to probably test this class and some more words just in case"))
print(sf.sortWordsbySize("Test line of text to probably test this class and some more words just in case"))
print(sf.sortWordsbyAlphabet("Test line of text to probably test this class and some more words just in case"))
        
```
